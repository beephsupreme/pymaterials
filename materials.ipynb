```python
# -*- coding: utf-8 -*-
from IPython.display import display
import inventory
import shipping
import sales
import report
import pandas as pd
import constants as cn


def run():
    print("Loading files ...")
    data = inventory.build()
    hfr = sales.build(cn.HFR_EXPORT)
    backlog = sales.build(cn.BACKLOG_EXPORT)
    schedule = shipping.build(data[cn.PN].tolist())
    df = report.build(data, schedule, backlog, hfr)
    df.style.set_caption("Shit Howdy").format(precision=0).background_gradient()
    display(df)
    print("Finished!")
```

```python
pd.set_option('display.max_rows', None)
run()
```
